```python
from sklearn.preprocessing import LabelEncoder
Label_encoder = LabelEncoder()
```

```python
df.salary = Label_encoder.fit_transform(df['salary'])
df.Department = Label_encoder.fit_transform(df['Department'])
```

```python
df.head()
```

```python
x = df.drop('left',axis = 1)
y = df['left']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2,random_state=40)
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
```

```python
xtrain_scaled = std_scaler.fit_transform(x_train)
xtest_scaled = std_scaler.fit_transform(x_test)
```

```python
xtrain_scaled
xtest_scaled
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest_model = RandomForestClassifier()
```

```python
random_forest_model.fit(xtrain_scaled,y_train)
```

```python
y_pred = random_forest_model.predict(x_test) 
```

```python
y_pred = random_forest_model.predict(xtrain_scaled)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
sns.heatmap(cm,annot=True,fmt = 'd')
```

```python
from sklearn.metrics import accuracy_score
model_accuracy = accuracy_score(y_test,y_pred)
print(model_accuracy)
```

```python
from sklearn.metrics import precision_score
model_precision = precision_score(y_test,y_pred)
print(model_precision)
```

```python
from sklearn.metrics import recall_score
model_recall = recall_score(y_test,y_pred)
print(model_recall)
```

```python
from sklearn.metrics import f1_score
model_f1 = f1_score(y_test,y_pred)
print(model_f1)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
score_list = random_forest_model.feature_importances_
list_of_features = list(x.columns)
print(len(score_list), len(list_of_features))

scores_df = pd.DataFrame({'scores': score_list, 'feature': list_of_features})
scores_df.sort_values(by='scores', ascending=False, inplace=True)

print(scores_df)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Feature names and importances
features = list(x.columns)
importances = random_forest_model.feature_importances_

plt.figure(figsize=(8, 6))
plt.barh(range(len(features)), importances)
plt.yticks(range(len(features)), features)
plt.xlabel('Importance Score')
plt.ylabel('Features')
plt.title('Feature Importances')
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(random_forest_model,xtrain_scaled,y_train,cv=5,scoring = 'accuracy')
print(scores)
```

```python
Avg_model_score = scores.mean()
print(Avg_model_score)
```
